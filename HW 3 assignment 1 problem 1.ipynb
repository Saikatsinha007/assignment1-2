```python

unique_players = df['Player'].unique()
highest_rank_years = {}

for player in unique_players:
    player_data = df[df['Player'] == player]
    player_rank_by_year = player_data.groupby('YEAR')['Rank'].min().reset_index()
    
    if not player_rank_by_year.empty:
        highest_rank_year = player_rank_by_year.loc[player_rank_by_year['Rank'].idxmin(), 'YEAR']
        highest_rank_years[player] = highest_rank_year
    else:
      highest_rank_years[player] = None
for player, year in highest_rank_years.items():
    print(f"Year with the highest rank for {player}: {year}")
```

# Average runs per match

```python
 unique_players = df['Player'].unique()
unique_years = df['YEAR'].unique()
average_runs_per_match_by_player_and_year = {}

for player in unique_players:
    player_data = df[df['Player'] == player]
    average_runs_per_match_by_player_and_year[player] = {}
    for year in unique_years:
        player_data_year = player_data[player_data['YEAR'] == year]
        player_runs_year = player_data_year['Runs'].sum()
        player_matches_year = player_data_year['Matches'].sum()
        if player_matches_year != 0:
            average_runs_per_match_year = player_runs_year / player_matches_year
        else:
            average_runs_per_match_year = 0
        average_runs_per_match_by_player_and_year[player][year] = average_runs_per_match_year

for player, data in average_runs_per_match_by_player_and_year.items():
    print(f"Player: {player}")
    for year, average_runs_per_match_year in data.items():
        print(f"  - Year: {year}, Average runs per match: {average_runs_per_match_year}")
```

```python
unique_players = df['Player'].unique()
unique_years = df['YEAR'].unique()
average_runs_per_match_by_player_and_year = {}
total_runs_by_player = {}
total_matches_by_player = {}

for player in unique_players:
    player_data = df[df['Player'] == player]
    average_runs_per_match_by_player_and_year[player] = {}
    total_runs_by_player[player] = 0
    total_matches_by_player[player] = 0
    for year in unique_years:
        player_data_year = player_data[player_data['YEAR'] == year]
        player_runs_year = player_data_year['Runs'].sum()
        player_matches_year = player_data_year['Matches'].sum()
        total_runs_by_player[player] += player_runs_year
        total_matches_by_player[player] += player_matches_year
        if player_matches_year != 0:
            average_runs_per_match_year = player_runs_year / player_matches_year
        else:
            average_runs_per_match_year = 0
        average_runs_per_match_by_player_and_year[player][year] = average_runs_per_match_year

for player, data in average_runs_per_match_by_player_and_year.items():
    print(f"Player: {player}")
    for year, average_runs_per_match_year in data.items():
        print(f"  - Year: {year}, Average runs per match: {average_runs_per_match_year}")
    print(f"  ----------Total Runs: {total_runs_by_player[player]}, Total Matches: {total_matches_by_player[player]}, Avg_run_rate: {total_runs_by_player[player]/total_matches_by_player[player]}")
```
